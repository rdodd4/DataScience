```python
from sklearn.datasets import fetch_20newsgroups
categories = ['sci.space']
dataset = fetch_20newsgroups(subset='all', shuffle=True, random_state=42, categories=categories)
corpus=dataset.data
```

```python
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
nltk.download('stopwords')
```

```python
corpus
```

```python
corpus = [x.lower() for x in corpus]
```

```python
stopset = set(stopwords.words('english'))
```

```python
stopset.update(['\n','\n>','\n>>','profit','fee','funny','overhead','husband','poison','SunOS','budget','>\t','unnecessary',
                'breathind','comparision','quick','editor','approved','permit','Smith','country','political','JPL','Jack Coyote'
               ,'David McAloon','business','Ken','mirror','wires','\t','strange','volume','decayed','sinking','Accomplised'
               ,'point','contacts','Durand','charts','record','experiments',])
```

```python
corpus[0]
```

```python
vectorizer = TfidfVectorizer(stop_words=stopset, use_idf= True, ngram_range= (1,3))
Y = vectorizer.fit_transform(corpus)
```

```python
Y[0]
```

```python
print Y[0]
```

```python
Y.shape
```

```python
lsa = TruncatedSVD(n_components=27, n_iter=100)
lsa.fit(Y)
```

```python
lsa.components_[0]
```

```python
terms = vectorizer.get_feature_names()
for i, comp in enumerate(lsa.components_): 
    termsInComp = zip (terms,comp)
    sortedTerms =  sorted(termsInComp, key=lambda x: x[1], reverse=True) [:10]
    print ( "Concept %d:" % i )
    for term in sortedTerms:
        print (term[0])
    print (" ")
```
